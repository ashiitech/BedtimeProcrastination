# Extracting values and categorizing

```python
for i in df["Bedtime"]:
    j=i.ljust(2)[:2].strip()
    df["Bedtime"].replace(i,j,inplace=True)
df['Bedtime'] = df['Bedtime'].str.replace('this_string', '')
df['Bedtime'] = df['Bedtime'].str.replace('\D', '', regex=True)
df['Bedtime'] = pd.to_numeric(df['Bedtime'], errors='coerce').fillna(0).astype(int)
df["Bedtime"]
```

```python
for i in df["Sleep Duration"]:
    j=i.ljust(2)[:2].strip()
    df["Sleep Duration"].replace(i,j,inplace=True)
df['Sleep Duration'] = df['Sleep Duration'].str.replace('this_string', '')
df['Sleep Duration'] = df['Sleep Duration'].str.replace('\D', '', regex=True)
df['Sleep Duration'] = pd.to_numeric(df['Sleep Duration'], errors='coerce').fillna(0).astype(int)
df['Sleep Duration']
```

# Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier


X = data[["Gender","Screen Time","Household Factors","Screen Time Duration","Awake-Experience","Sleep-Related Worries","Daytime Naps",
        "Entertainment","Bedtime","Pre-Sleep Social Media Use","Physical Activity","Bedtime Motivation","Sleep Quality Rating","Sleep Duration"]]
y = data['Age']


rf = RandomForestClassifier(n_estimators=100, random_state=0)
rf.fit(X, y)


importances = rf.feature_importances_


feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})


feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


top_n = 7

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'][:top_n], feature_importance_df['Importance'][:top_n], color='teal')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title(f'Top {top_n} Feature Importances')
plt.show()


print(f"Top {top_n} Feature Importances:")
print(feature_importance_df[['Feature', 'Importance']][:top_n])
```

```python
from sklearn.feature_selection import mutual_info_classif
%matplotlib inline
importances=mutual_info_classif(X,y)
feat_importances=pd.Series(importances,X.columns[0:len(X.columns)])
feat_importances.plot(kind='barh',color='teal')
plt.show()

threshold = 0.05

selected_features = feat_importances[feat_importances > threshold].index

print("Selected Features =",selected_features)
```

# K Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


X = data[['Age','Screen Time']]

silhouette_scores = []

k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, cluster_labels))

optimal_k = 10

final_kmeans = KMeans(n_clusters=optimal_k, random_state=0, n_init=10)
final_kmeans.fit(X)


```

# 7.Predict and Forecast And 8.Monitor and Iterate

```python
print(f"Optimal K: {optimal_k}")
sil = silhouette_score(X, cluster_labels)
print("Silhouette Score: ",sil)
```
